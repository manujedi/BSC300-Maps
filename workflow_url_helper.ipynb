```python
import requests
from bs4 import BeautifulSoup

base_url = "https://download.geofabrik.de/europe/"
country = "france"
response = requests.get(base_url + country + ".html")
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find_all('table', id="subregions")[1]
rows = table.find_all('tr')
```

```python
state_cnt = 1
for row in rows:
    for entry in row.find_all('a', href=True):
        link = entry['href']
        if "latest.osm.pbf" in link:
            name = link.split('/')[-1].replace('-latest.osm.pbf', '')
            print("{ name:", name,", geofabrik-url: ",base_url + link, ", state: \"" + str(state_cnt).rjust(2, '0') + "00\"},")
            state_cnt += 1
```
