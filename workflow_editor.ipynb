```python
from pathlib import Path
import pprint
import yaml
```

```python
# read current database

database = []

files = [f for f in Path("./.github").rglob("*.yml") if f.is_file()]
#files = [f for f in files if "Poland" in f.name]
files =  sorted(files, key=lambda x: x.name)
data = []

for i in files:
    if not i.name.startswith('build'):
        continue

    content = yaml.safe_load(open(i))

    # single country
    if 'url' in content['env'].keys():
        # database.append({'name': content['name'], 'url': content['env']['url'], 'countrycode':content['env']['countrycode']})
        database.append({'name': content['name'], 'maps': [{'name': content['name'], "url": content['env']['url'], "state": "0000"}], 'countrycode':content['env']['countrycode']})

    # country split in states
    if 'strategy' in content['jobs']['build'].keys():
        database.append({'name': content['name'], 'maps': content['jobs']['build']['strategy']['matrix']['maps'], 'countrycode':content['env']['countrycode']})

# fix norway (NO is false in yaml...)
for entry in database:
    if entry['name'] == 'norway':
        entry['countrycode'] = 'NO'
pprint.pp(database)
```

```python
# write database to files DO NOT USE YAML lib

for entry in database:
    # single country
    if 'url' in entry.keys():
        src = open('template_single_country.yml', 'r')
        dst = open('.github/workflows/build-' + entry['name'] + '.yml', 'w')
        for index, line in enumerate(src):
            if index == 0:
                dst.write('name: ' + entry['name'] + '\n')
            elif index == 3:
                dst.write('  url: ' + entry['url'] + '\n')
            elif index == 4:
                dst.write('  countrycode: ' + entry['countrycode'] + '\n')
            else:
                dst.write(line)
        dst.close()

    if 'maps' in entry.keys():
        src = open('template_state_country.yml', 'r')
        dst = open('.github/workflows/build-' + entry['name'] + '.yml', 'w')
        for index, line in enumerate(src):
            if index == 0:
                dst.write('name: ' + entry['name'] + '\n')
            elif index == 3:
                dst.write('  countrycode: ' + entry['countrycode'] + '\n')
            elif index == 19:
                dst.write('          [\n')
                for map in entry['maps']:
                    dst.write('            {name: ' + map['name'] + ', url: ' + map['url'] + ', state: "' + map['state'] +'"},\n')
                dst.write('          ]\n')
            else:
                dst.write(line)
        dst.close()
```
