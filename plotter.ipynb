```python
import pandas as pd
```

```python
df = pd.read_csv('states.csv')
df['country']=df['map'].str[:2]
df['state']=df['map'].str[2:6]
df['date']=df['map'].str[6:12]
df['west']=df['map'].str[12:15]
df['north']=df['map'].str[15:18]
df['west-east']=df['map'].str[18:21]
df['north-south']=df['map'].str[21:24]
```

```python
for index, row in df.iterrows():
  df.loc[index, 'west'] = int(row['west'], base=36)
  df.loc[index, 'north'] = int(row['north'], base=36)
  df.loc[index, 'west-east'] = int(row['west-east'], base=36)
  df.loc[index, 'north-south'] = int(row['north-south'], base=36)
```

```python

df=df.drop(columns=['map','country','state','date','west-east', 'north-south'])
df
```
